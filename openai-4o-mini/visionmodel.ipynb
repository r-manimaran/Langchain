```python
from openai import OpenAI
```

```python
model_name="gpt-4o-mini"
```

```python
client = OpenAI()
completion = client.chat.completions.create(
  model=model_name,
  messages=[
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "Hello!"}
  ]
)

print(completion.choices[0].message.content)
```

```python
completion = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Write 5 dictionary words which starts with E and ends with N"}
    ]
)
print(completion.choices[0].message.content)
```

```python
from IPython.display import Image
image_path = "architecture.png"
display(Image(image_path))
```

```python
import base64
def encode_image(image):
    with open(image, 'rb') as image_file:
        encoded_image = base64.b64encode(image_file.read()).decode('utf-8')
    return encoded_image
```

```python
base64_image = encode_image(image_path)
```

```python
response = client.chat.completions.create(
    model = model_name,
    messages =[
        {"role":"system","content":"You are a helpful assistant that responds in markdown. Help me to answer to the question"},
        {"role":"user","content":[
            {"type":"text","text":"can you explain this architecure diagram"},
            {"type":"image_url","image_url":{"url":f"data:image/png;base64,{base64_image}"}}
        ]}
    ],
    temperature=0.2
)
```

```python
print(response.choices[0].message.content)
```
